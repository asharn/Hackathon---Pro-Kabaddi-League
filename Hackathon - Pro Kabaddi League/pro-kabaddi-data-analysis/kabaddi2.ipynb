```python
# Any results you write to the current directory are saved as output.
def flatten_json(data, delim ):
    val = {}
    if isinstance(data, dict ):
        for i in data.keys():
            if isinstance(data[i], dict ):
                get = flatten_json( data[i], delim )
                for j in get.keys():
                    val[ i + delim + j ] = get[j]
            elif isinstance( data[i], list):
                for j in data[i]:
                    get = flatten_json( j, delim )
                    for k in get.keys():
                        val[ i + delim + k ] = get[k]
            else:
                val[i] = data[i]
    #else:
    #    return val[data] = dict(data)
    return val
#print(pd.__version__)
```

# 1. Data Cleaning and manipulation

```python
# getting total number of rows and column in the dataframe
print(f" Shape of the dataframe = {df3.shape}")
totalrows=df3.shape[0]
print(f" Total number of rows in the dataset =  {totalrows}")
```

```python
#Since we have 145 columns lets remove the columns which have more than 50% null values so that we can reduce the columns to be used in analysis
df_nullcols = df3.isnull().sum()
df_nullcols = df_nullcols[df_nullcols.values >(0.5*totalrows)]

#Columns which has more than 50% null values and we can remove them from the data frame
df_columnstobedropped=df_nullcols.index
```

```python
# This has the list of the columns which will be dropped as it has more than 50% of the values as null values
list(df_columnstobedropped)
#111-57
```

```python
# function to drop columns which is having more than 50%  values as null.
def dropcolumns(columnlist):
    for col in columnlist:
        df3.drop(col,axis=1,inplace=True)
```

```python
# Dropping columns  which is having more than 50%  values as null.
dropcolumns(df_columnstobedropped)
```

```python
# after dropping the required columns we have 136 columns in the data
df3.shape
```

```python
# Identifying the columns having high nan after all the above data cleansing. For Employee title and Employee_Length nan can be replaced by unknown
filter = df3[:] != " "
dfNew = df3[filter]
```

```python
# dropping colunm with more than 50%  emmpty values

df3 = df3.dropna(thresh=int(df3.shape[0] * .5), axis=1)
```

```python
#nunique() gets count of unique values for every col on the data frame

df_unqvalcols=df3.nunique()
df_dropcollist=df_unqvalcols[(df_unqvalcols.values==1) ]
#| (df_unqvalcols.values==totalrows)
list(df_dropcollist.index)
df3.drop(labels=list(df_dropcollist.index),axis=1,inplace=True)

df3.shape
```

```python
# Replacing empty value with median
df3[:] = df3[:].fillna(df3[:].median())
```
